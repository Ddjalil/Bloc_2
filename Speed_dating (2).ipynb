```python
only_id["gender"] = dataset["gender"].apply(lambda x: "Woman" if x == 0 else "Man" if x ==1 else x)
```

```python
only_id['match'] = only_id['match'].apply(lambda x: 'Match' if x == 1 else 'No match')
```

```python
matches_nbre = only_id['match'].value_counts()

percent_match = matches_nbre/len(dataset)*100
```

```python
# Visualisation des "match" en %
fig, ax = plt.subplots(figsize = (8,5))
ax.bar( percent_match.index, percent_match.values )
plt.title(" Pourcentage des matchs ")
plt.show()
```

```python

career_interest_counts = only_id['goal'].value_counts()

# Définir un dictionnaire pour les correspondances des valeurs dans la légende
labels_dict = {
    1: 'Fun night',
    2: 'Meet new people',
    3: 'Get a date',
    4: 'Serious relationship',
    5: 'I did it',
    6: 'Other',
}

# Mapper les légendes avec le dictionnaire de correspondance
labels = [labels_dict.get(label, label) for label in career_interest_counts.index]

# Créer le graphique en camembert avec les légendes personnalisées
plt.pie(career_interest_counts, labels=labels, autopct='%1.1f%%')

# Ajouter un titre au graphique
plt.title('Objectif principal des participants')

# Afficher le graphique
plt.show()
```

```python
# Définir un dictionnaire de correspondance pour les métiers
metiers_dict = {
    1: 'Lawyer',
    2: 'Academic',
    3: 'Psychologist',
    4: 'Doctor',
    5: 'Engineer',
    6: 'Arts',
    7: 'Finance',
    8: 'Estate',
    9: 'Humanitarian',
    10: 'Undecided',
    11: 'Social',
    12: 'Pathology',
}
# Grouper les participants par métier et compter le nombre de participants dans chaque métier
career_counts = only_id['career_c'].map(metiers_dict).value_counts().reset_index()

# Renommer les colonnes
career_counts.columns = ['Métier', 'Nombre de participants']

# Créer le graphique interactif avec Plotly
fig = px.bar(career_counts, x='Métier', y='Nombre de participants')

# Personnaliser le titre du graphique
fig.update_layout(title='Répartition des participants par métier')

# Afficher le graphique interactif
fig.show()
```

```python
# Variation de l'attractivité des participants en fonction de l'âge :

age_attr_mean = only_id.groupby('age')['attr1_1'].mean()
plt.plot(age_attr_mean.index, age_attr_mean.values)
plt.xlabel('Âge')
plt.ylabel('Attractivité moyenne')
plt.title('Variation de l\'attractivité des participants en fonction de l\'âge')
plt.show()
```

```python
# Distribution des scores de sincérité des participants :

plt.hist(only_id['sinc1_1'], bins=20)
plt.xlabel('Score de sincérité')
plt.ylabel('Nombre de participants')
plt.title('Distribution des scores de sincérité des participants')
plt.show()
```

```python
# Corrélation entre l'âge des participants et l'intérêt pour une relation à long terme :

plt.scatter(only_id['age'], dataset['long_term'])
plt.xlabel('Âge')
plt.ylabel('Intérêt pour une relation à long terme')
plt.title('Corrélation entre l\'âge et l\'intérêt pour une relation à long terme')
plt.show()
```

```python
# Taux de décision positive des partenaires en fonction de l'intérêt pour la musique :

music_interest_dec_rates = only_id.groupby('music')['dec_o'].mean()
plt.plot(music_interest_dec_rates.index, music_interest_dec_rates.values)
plt.xlabel('Intérêt pour la musique')
plt.ylabel('Taux de décision positive')
plt.title('Taux de décision positive des partenaires en fonction de l\'intérêt pour la musique')
plt.show()
```

```python
# Répartition des préférences religieuses des participants :


religion_counts = only_id['imprelig'].value_counts()
plt.pie(religion_counts, labels=religion_counts.index, autopct='%1.1f%%')
plt.title("Note sur 10 de l 'importance religieuses des participants")
plt.show()
```

```python
# Comparaison de l'intérêt pour les sports entre hommes et femmes : 



gender_sports_interest = only_id.groupby('gender')['sports'].mean()
plt.bar(gender_sports_interest.index, gender_sports_interest.values)
plt.xlabel('WOMAN                       MAN')
plt.ylabel('Intérêt pour les sports')
plt.title('Comparaison de l\'intérêt pour les sports entre hommes et femmes')
plt.show()
```
